# Applied Algorithms - String Permutations

```python
def permutation(input_string):
    #list to store the permutations
    output_strings = set()
    
    #base condition
    if len(input_string)<2:
        return input_string
    
    #Recursive condition
    for i in range(0,len(input_string)):
        temp_string = input_string[0:i] + input_string[i+1:]
        #Fixing one letter and then making recursive call for letters other than the fixed one
        for j in permutation(temp_string):
            #Appending the result to the final list
            output_strings.add(input_string[i]+j)

    return output_strings
```
